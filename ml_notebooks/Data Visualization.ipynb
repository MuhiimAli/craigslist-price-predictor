```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
with open("../data/item_with_images.json", "r") as items_json:
    items = json.load(items_json)
```

```python
items_df = pd.DataFrame(items)
```

```python
items_df.head()
```

```python
items_df = items_df[items_df["category"] != "for"]
```

```python
cat_map = {
    "spo": "Sporting Goods",
    "clt": "Collectibles",
    "tls": "Tools",
    "fuo": "Furniture",
    "hsh": "Household",
    "pts": "Auto Parts",
    "grd": "Garden",
    "msg": "Music Instruments",
    "ctd": "Cars and Trucks",
    "clo": "Clothes and Accessories"
}
cat_map.values()
```

```python
plt.figure(figsize = (12, 10))
sns.countplot(data = items_df, x = "category", order = items_df['category'].value_counts().index[:10])
plt.title("10 Most Common Categories")
plt.xticks(ticks = [i for i in range(10)], labels = list(cat_map.values()), rotation = -45, ha='left')
plt.subplots_adjust(bottom=0.3)
plt.savefig("../analysis_deliverable/visualizations/common_categories.png")
pass
```
